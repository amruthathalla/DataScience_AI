```python
#Outer loop (i) runs 5 times, For each outer loop iteration:, Prints data science once, Inner loop (j) runs 4 times, Prints technology 4 times, Prints a blank line, Then moves to the next i
i = 1

while i<=5:
    print('data science') # when we mention end then new line will not create
    j = 1
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()
```

```python
#Outer loop (i) runs 5 times, For each outer iteration:, datascience prints once, Inner loop (j) prints technology 4 times on the same line, print() moves to the next line
i = 1
while i<=5:
    print(' datascience', end = "") # when we mention end then new line will not create
    j = 1
    while j<=4:
        print(' technology', end="")
        j = j + 1
        
    i = i + 1
    print()
```

```python
#Outer loop (i) runs for: 1, 2., Inner loop (j) runs for: 0, 1, 2, Each row prints values of i * j, end=" " keeps values on the same line, print() moves to the next row
i = 1

while i <= 2 :
    j = 0
    while  j <= 2 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1 
```

```python
#Outer loop i → 1, 2, 3, 4, Inner loop j → 0, 1, 2, 3, Each row prints values of i * j
i = 1
while i <= 4 :
    j = 0
    while  j <= 3 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1 
```

Conditional Statements (Decision Making)

Conditional statements let your program make decisions and run different code depending on whether a condition is True or False.

Think of it like:

If it rains → take umbrella
Else → go without umbrella

1. if Statement

Runs code only if condition is True

Syntax
if condition:
    code

```python
age = 20

if age >= 18:
    print("Eligible to vote")
```

2. if–else Statement

Runs one block if True, another if False.

```python
num = 7

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

3. if–elif–else Ladder

Used when checking multiple conditions.

Python checks from top → bottom and stops at first True condition.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

4. Nested if

An if inside another if.

```python
age = 25
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
```

Short-Hand if (Ternary Operator)
One-line if–else.

```python
num = 10
result = "Even" if num % 2 == 0 else "Odd"
print(result)
```

Doing repeated tasks

Python has two main loops:

while loop

while Loop

Runs while condition is True.

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

Infinite Loop 
This never stops unless manually broken.

```python
while True:
    print("Running")
```

while with else

else runs when loop finishes normally (not by break).

```python
count = 1

while count <= 3:
    print(count)
    count += 1
else:
    print("Completed")
```

Used to iterate over collections (list, string, range, dictionary, set).

```python
for i in range(5):
    print(i)
```

Reverse for Loop

```python
for i in range(5, 0, -1):
    print(i)
```

Loop Through List

```python
for x in [10, 20, 30]:
    print(x)
```

Loop Through String

```python
for ch in "Python":
    print(ch)
```

Loop Through Dictionary

```python
student = {"name": "Afnan", "age": 22}

for key, value in student.items():
    print(key, value)
```

Loop Control Statements

These change loop behavior.
break → Stop Loop Immediately

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

continue → Skip Current Iteration

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

pass → Do Nothing Placeholder

Used when syntax requires code but you don’t want action yet.

```python
for i in range(3):
    pass
```

Nested Loops

Loop inside another loop.

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```
