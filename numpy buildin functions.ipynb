```python
#Array manipulation functions
```

```python
#reshape an array

a1=np.array([1,2,3])
reshaped=np.reshape(a1,(3,1))
print("reshaped arrays:",reshaped)
```

```python
#flatten an array -- 

f1=np.array([[1,2],[3,4]])
flattened = np.ravel(f1) #flatten to 1d array
print("flattened array:",flattened)
```

```python
#transpose an array --- reverse rows to colums and cols to rows

e1=np.array([[1,2],[3,4]])
transposed = np.transpose(e1)
print("transposed array:\n", transposed)
```

```python
# stack arrays vertically

a2=np.array([1,2])
b2=np.array([3,4])
stacked = np.vstack([a2,b2])
stacked
```

```python
# stack arrays horizontally

a2=np.array([1,2])
b2=np.array([3,4])
stacked = np.hstack([a2,b2])
stacked
```

```python
#Set operations
```

```python
#intersection
set_a=np.array([1,2,3,4])
set_b=np.array([3,4,5,6])

intersection = np.intersect1d(set_a,set_b)
print(intersection)
```

```python
union=np.union1d(set_a,set_b)
print(union)
```

```python
#Array Attributes functions
```

```python
a = np.array([1, 2, 3])
shape = a.shape  # Shape of the array
size = a.size    # Number of elements
dimensions = a.ndim  # Number of dimensions
dtype = a.dtype   # Data type of the array

print("Shape of a:", shape)
print("Size of a:", size)
print("Number of dimensions of a:", dimensions)
print("Data type of a:", dtype)
```
